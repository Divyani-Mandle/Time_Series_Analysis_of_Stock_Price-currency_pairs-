```python
EURUSD_df['returns'] = (EURUSD_df['Close']/EURUSD_df['Close'].shift(1))-1
USDJPY_df['returns'] = (USDJPY_df['Close']/USDJPY_df['Close'].shift(1))-1
EURJPY_df['returns'] = (EURJPY_df['Close']/EURJPY_df['Close'].shift(1))-1
```

```python
#higher the standard deviation is the higher the variance is so the more volatility we have in terms of stock
#EURJPY is more Volatile and had more fluctuation in prices as compare to EURUSD & USDJPY.
EURUSD_df['returns'].hist(bins=100, label ='EURUSD_df', alpha=0.5, figsize=(13,6))
USDJPY_df['returns'].hist(bins=100, label ='USDJPY_df',alpha=0.5)
EURJPY_df['returns'].hist(bins=100, label ='EURJPY_df',alpha=0.5)
plt.legend()
```

```python
EURUSD_df['returns'].plot(kind='kde',label='EURUSD_df', figsize=(15,6))
USDJPY_df['returns'].plot(kind='kde',label='USDJPY_df')
EURJPY_df['returns'].plot(kind='kde',label='EURJPY_df')
plt.legend()
```

```python
box_df=pd.concat([EURUSD_df['returns'],USDJPY_df['returns'],EURJPY_df['returns']],axis=1)
box_df.columns=['EURUSD returns','USDJPY returns','EURJPY returns']
box_df.plot(kind='box', figsize=(16,6))
#EURJPY seems more volatile than EURUSD &USDJPY
```

```python
scatter_matrix(box_df, figsize=(10,10),hist_kwds={'bins':50}, alpha =0.25)
```

```python
#cumulative return 
EURUSD_df['Cumulative returns'] = (1 + EURUSD_df['returns']).cumprod()
USDJPY_df['Cumulative returns'] = (1 + USDJPY_df['returns']).cumprod()
EURJPY_df['Cumulative returns'] = (1 + EURJPY_df['returns']).cumprod()
```
